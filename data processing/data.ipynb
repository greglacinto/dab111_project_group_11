```python
import pandas as pd
customer_data = pd.read_csv("https://download.data.world/s/5jl6yrklk6fpswmbwbp2og5wfe4erd?dws=00000")
```

```python
#dropping columns
customer_data.drop("is_active_member",axis=1,inplace=True)
customer_data.drop("has_creditcard",axis=1,inplace=True)
```

```python
# creating new invoice column and changing it to a string
customer_data["invoice_id"]=customer_data["invoice_id"].astype("string")
customer_data["branch"]=customer_data["branch"].astype("string")
customer_data["gender"]=customer_data["gender"].astype("string")
customer_data["customer_type"]=customer_data["customer_type"].astype("string")
customer_data["product_category"]=customer_data["product_category"].astype("string")
customer_data["customer_churn"]=customer_data["customer_churn"].astype("bool")

customer_data.head(10)
```

```python
import sqlite3

con = sqlite3.connect("../database/customers.db")
cur = con.cursor()

customer_data.to_sql("customer_data",con,index=False)

rows = cur.execute("SELECT * FROM customer_data").fetchall()
rows

con.commit()
con.close()
```
